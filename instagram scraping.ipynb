```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import requests
import os
import time

# Credentials (Use secure methods for real-world apps)
INSTAGRAM_USER = ""
INSTAGRAM_PASS = ""

# Setup WebDriver and login
driver = webdriver.Chrome()
driver.get("https://www.instagram.com/accounts/login/")
WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.NAME, "username")))

# Input credentials and login
driver.find_element(By.NAME, "username").send_keys(INSTAGRAM_USER)
driver.find_element(By.NAME, "password").send_keys(INSTAGRAM_PASS)
driver.find_element(By.CSS_SELECTOR, "button[type='submit']").click()

# Wait for home page to load, then go to profile
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "svg[aria-label='Home']")))
driver.get("https://www.instagram.com/propertiesbyshola/")

# Scroll page to load more content
for _ in range(5):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

# Parse the page with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')
post_divs = soup.find_all('div', class_='_aagu')

# List of unwanted fragments and patterns to be cleaned
unwanted_fragments = [
    '<div class="_aagu">', '<div class="_aagv">', '<img alt="', '<div class="_aagw">', 
    '</div>', 'class="x5yr21d xu96u03 x10l6tqk x13vifvy x87ps6o xh8yej3"', 
    'crossorigin="anonymous"', 'src', 'style="object-fit: cover;"/>', 
]

# Clean and print the posts
for idx, post in enumerate(post_divs):
    raw_text = post.prettify()
    
    # Remove unwanted fragments
    for fragment in unwanted_fragments:
        raw_text = raw_text.replace(fragment, '')
    
    # Exclude any line containing a URL pattern (like instagram.fabv)
    cleaned_text = "\n".join([line for line in raw_text.splitlines() if "https://instagram.fabv" not in line])
    
    # Print the cleaned post
    print(f"Post {idx + 1}:", cleaned_text)

# Close the browser
driver.quit()
```
